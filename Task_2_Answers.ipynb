# Question 3

```python
columns_to_calculate = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

# Calculate mean, standard deviation, and quartiles
mean_values = df[columns_to_calculate].mean()
std_values = df[columns_to_calculate].std()
quartile_values = df[columns_to_calculate].quantile([0.25, 0.5, 0.75])

# Combine the results into a new DataFrame
summary_df = pd.DataFrame({
    'Mean': mean_values,
    'Standard Deviation': std_values,
    '25th Percentile': quartile_values.loc[0.25],
    '50th Percentile (Median)': quartile_values.loc[0.5],
    '75th Percentile': quartile_values.loc[0.75]
})

# Display the summary DataFrame
summary_df
```

# Question 4

```python
import matplotlib.pyplot as plt

# Plot histograms in a 1x4 layout
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for i, column in enumerate(columns_to_calculate):
    ax = axes[i]
    df[column].plot(kind='hist', ax=ax, title=column)
    ax.set_xlabel(column)
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# Question 5

```python
import seaborn as sns

# Visualize scatter plots and histograms
sns.pairplot(df[columns_to_calculate])
plt.show()
```

# Question 6

```python
selected_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
new_df = df[selected_columns]

# Create a PairGrid with customized plots
g = sns.PairGrid(new_df)

# Plots on the diagonal
g.map_diag(sns.histplot, kde=True)

# Plots on the upper triangle
g.map_upper(sns.histplot, bins=10)  # Bivariate histogram
g.map_upper(sns.kdeplot, cmap='Blues_d')  # Bivariate KDE plot

# Plots on the lower triangle
g.map_lower(sns.kdeplot, cmap='Reds_d')  # Bivariate KDE plot

plt.show()
```
