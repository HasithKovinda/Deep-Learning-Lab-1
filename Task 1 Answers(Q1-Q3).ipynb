### **Question 2**

```python
import matplotlib.pyplot as plt

# Generate a random 100000x1 array from an exponential distribution
random_exponential = np.random.exponential(scale=1, size=100000)

# Create histograms for different distributions
plt.hist(random_exponential, density=True, bins=50, histtype="step", color="blue", label="Exponential")
plt.hist(np.random.rand(100000), density=True, bins=50, histtype="step", color="orange", label="Uniform")
plt.hist(np.random.randn(100000), density=True, bins=50, histtype="step", color="green", label="Normal")

# Set plot properties
plt.legend()
plt.title("Random Distributions Comparison")
plt.xlabel("Value")
plt.ylabel("Density")
plt.xlim(-2, 10)  # Adjust the x-axis limit for better visualization

plt.show()
```

### **Question 3**

```python
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate the Z values using the function Z = X^2 + Y^2
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plot of Z = X^2 + Y^2')

# Show the plot
plt.show()
```
